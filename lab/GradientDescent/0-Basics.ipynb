#梯度下降法
## 一、 什么事梯度下降法
- 不是一个机器学习算法
- 是一种基于搜索的优化方法
- 作用：最小化一个损失函数
- 梯度上升法：最大化一个效用函数
- 并不是所有函数都有唯一的极值点
    > 解决方案：
    - 多次运行，随机化初始点
    - 梯度下降法的初始点也是一个超参数

eg: 线性回归中使用梯度下降法

线性回归中我们的目标是使得$ \sum_{i=1}^m(y^i - \bar y^i)^2 $尽可能小

所以线性回归的损失函数是具有唯一的最优解

# 梯度下降法模拟

```python
import numpy as np
import matplotlib.pyplot as plt

plot_x = np.linspace(-1,6,141)
plot_y = (plot_x - 2.5) ** 2 - 1
plt.plot(plot_x,plot_y)
plt.show()
```

```python
def dJ(theta):
    return 2 * (theta - 2.5)


def J(theta):
    return (theta - 2.5) ** 2 - 1
```

```python
theta = .0
eta = .1
epsilon = 1e-8
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    if (abs(J(theta) - J(last_theta))) < epsilon:
        break
plt.plot(plot_x,J(plot_x))
plt.plot(np.array(theta_history),J(np.array(theta_history)),color="r",marker="*")
plt.show()

```
